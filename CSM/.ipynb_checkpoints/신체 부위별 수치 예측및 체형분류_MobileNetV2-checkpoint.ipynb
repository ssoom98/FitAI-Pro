```python
import tensorflow as tf
import numpy as np
import os
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder

# 이미지 경로 및 CSV 파일 설정
image_folder = "E:/image_data"
csv_file_path = "E:/body_shape.csv"

# CSV 데이터 로드
df = pd.read_csv(csv_file_path)

# 신체 둘레 데이터 컬럼 (정규화 필요)
body_measure_cols = ["height", "weight", "body_fat", "waist", "belly_waist", "hip", "thigh", "bust", "arm"]
df[body_measure_cols] = df[body_measure_cols].astype(np.float32)

# 체형 데이터 인코딩
label_encoder = LabelEncoder()
df["body_shape_encoded"] = label_encoder.fit_transform(df["body_shape"])

# 이미지 파일 경로 생성 함수
def get_image_paths(model_num):
    base_filenames = ["01", "03"]
    image_paths = []

    for base_filename in base_filenames:
        for i in range(16):  
            file_name = f"resize_{model_num}_{i+1}.jpg"
            file_path = os.path.join(image_folder, file_name)
            if os.path.exists(file_path):
                image_paths.append(file_path)

    return image_paths

# 이미지-라벨 데이터 매칭
image_label_pairs = []
for _, row in df.iterrows():
    model_num = row["model_num"]
    image_paths = get_image_paths(model_num)

    if image_paths:
        for img_path in image_paths:
            image_label_pairs.append((img_path, row[body_measure_cols].values, row["body_shape_encoded"]))

print(f"총 {len(image_label_pairs)}개 데이터 로드 완료")

# 데이터셋 분할 (Train 70%, Validation 15%, Test 15%)
def split_data(image_paths, body_measures, body_shapes):
    train_paths, temp_paths, train_measures, temp_measures, train_shapes, temp_shapes = train_test_split(
        image_paths, body_measures, body_shapes, test_size=0.3, random_state=42
    )

    val_paths, test_paths, val_measures, test_measures, val_shapes, test_shapes = train_test_split(
        temp_paths, temp_measures, temp_shapes, test_size=0.5, random_state=42
    )

    print(f"데이터 분할 완료 (Train: {len(train_paths)}, Val: {len(val_paths)}, Test: {len(test_paths)})")
    return train_paths, val_paths, test_paths, train_measures, val_measures, test_measures, train_shapes, val_shapes, test_shapes

# TFRecord 변환 함수
def serialize_example(image_path, body_measures, body_shape_label):
    image = tf.io.read_file(image_path)
    image = tf.image.decode_jpeg(image, channels=3)
    image = tf.image.resize(image, [224, 224])
    image = tf.image.convert_image_dtype(image, tf.float32)
    image = tf.io.encode_jpeg(tf.cast(image * 255, tf.uint8))

    feature = {
        "image": tf.train.Feature(bytes_list=tf.train.BytesList(value=[image.numpy()])),
        "body_measures": tf.train.Feature(float_list=tf.train.FloatList(value=body_measures)),
        "body_shape_label": tf.train.Feature(int64_list=tf.train.Int64List(value=[body_shape_label]))
    }

    example = tf.train.Example(features=tf.train.Features(feature=feature))
    return example.SerializeToString()

# TFRecord 저장 함수
def save_tfrecord(file_path, image_paths, body_measures, body_shapes):
    with tf.io.TFRecordWriter(file_path) as writer:
        for img_path, measures, shape in zip(image_paths, body_measures, body_shapes):
            tf_example = serialize_example(img_path, measures, shape)
            writer.write(tf_example)
    print(f"TFRecord 저장 완료: {file_path}")

# TFRecord 데이터 로딩 함수 + 데이터 증강 추가
def parse_tfrecord(example_proto):
    feature_description = {
        "image": tf.io.FixedLenFeature([], tf.string),
        "body_measures": tf.io.FixedLenFeature([9], tf.float32),
        "body_shape_label": tf.io.FixedLenFeature([], tf.int64),
    }

    parsed_features = tf.io.parse_single_example(example_proto, feature_description)

    image = tf.io.decode_jpeg(parsed_features["image"], channels=3)
    image = tf.image.convert_image_dtype(image, tf.float32)
    image = tf.image.resize(image, [224, 224])
    image = image / 256  # 정규화

    # 데이터 증강 적용 (train dataset만 적용)
    image = tf.image.random_flip_left_right(image)
    image = tf.image.random_brightness(image, 0.1)
    image = tf.image.random_contrast(image, 0.8, 1.2)
    image = tf.image.random_hue(image, 0.08)
    image = tf.image.random_saturation(image, 0.8, 1.2)

    body_measures = tf.cast(parsed_features["body_measures"], tf.float32)
    body_shape_label = tf.one_hot(parsed_features["body_shape_label"], depth=3)

    return image, (body_measures, body_shape_label)

# TFRecord 데이터 로드 함수
def load_tfrecord(file_path, batch_size=32):
    dataset = tf.data.TFRecordDataset(file_path)
    dataset = dataset.map(parse_tfrecord)
    dataset = dataset.shuffle(1000).batch(batch_size).prefetch(tf.data.experimental.AUTOTUNE)
    return dataset

# 전체 데이터셋 변환 및 저장 실행
image_paths, body_measures, body_shapes = zip(*image_label_pairs)

# 데이터셋 분할
train_paths, val_paths, test_paths, train_measures, val_measures, test_measures, train_shapes, val_shapes, test_shapes = split_data(
    image_paths, body_measures, body_shapes
)

# 저장할 TFRecord 파일 경로
train_tfrecord_file = "train_dataset.tfrecord"
val_tfrecord_file = "val_dataset.tfrecord"
test_tfrecord_file = "test_dataset.tfrecord"

# TFRecord 저장
save_tfrecord(train_tfrecord_file, train_paths, train_measures, train_shapes)
save_tfrecord(val_tfrecord_file, val_paths, val_measures, val_shapes)
save_tfrecord(test_tfrecord_file, test_paths, test_measures, test_shapes)

# 데이터셋 로드
train_dataset = load_tfrecord(train_tfrecord_file)
val_dataset = load_tfrecord(val_tfrecord_file)
test_dataset = load_tfrecord(test_tfrecord_file)

print("TFRecord 데이터셋 로딩 완료")
```
